```python
import os
from pathlib import Path
from dotenv import load_dotenv

from langchain_chroma import Chroma
from langchain_ollama import OllamaEmbeddings, ChatOllama
from langchain_core.tools import tool
from graph.consts import KB_PATH, EMBED_MODEL
from langchain_core.tools.retriever import create_retriever_tool

from langchain_ollama import ChatOllama
load_dotenv()

LLM_MODEL = "mistral:8b"

print(f"Loading Health Knowledge Base from: {KB_PATH}")
# Initialize local Embeddings
embeddings = OllamaEmbeddings(model=EMBED_MODEL)

# Load the Vector Store
vectorstore = Chroma(
    collection_name="health_knowledge",
    embedding_function=embeddings,
    persist_directory=str(KB_PATH)
)

retriever = Chroma(
    collection_name="health_knowledge",
    persist_directory=str(KB_PATH),
    embedding_function=embeddings,
).as_retriever()

print(f"Database active. Chunk count: {vectorstore._collection.count()}")
question = "what should i eat after sport?"
ret = retriever.invoke(question)
```
