```python
# cell 1
# pip install langchain langgraph langchain-openai

from langchain.agents import create_agent
from langchain_ollama import ChatOllama
from langchain_core.tools import tool
from langchain.chat_models import init_chat_model
```

```python
# cell 2
# Define two trivial tools

@tool
def add(a: int, b: int) -> int:
    """Add two numbers."""
    print(f"[TOOL] add({a}, {b})")
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    print(f"[TOOL] multiply({a}, {b})")
    return a * b
```

```python
# cell 3
# Create model and agent
model_id = "ollama:ministral-3:8b"
llm = init_chat_model(model=model_id, temperature=0)

agent = create_agent(
    model=llm,
    tools=[add, multiply],
    system_prompt=(
        "You are a reasoning agent.\n"
        "Use tools when calculation is needed.\n"
        "Think step by step and stop only when you are done."
    ),
)
```

```python
# cell 4
# Run with recursion limit to observe the loop

result = agent.invoke(
    {
        "messages": [
            {"role": "user", "content": "Compute (12 + 8) * 5 using tools"}
        ]
    },
    config={"recursion_limit": 10}
)

print("\nFINAL RESULT\n")
print(result)
```
